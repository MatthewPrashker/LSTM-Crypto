```python
import torch
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
def make_tensor(curr):
    data = pd.read_csv("Data/" + curr + "-USD.csv")
    
    #BTC time series data starts 242 days earlier than the others
    if curr == 'BTC':
        data = data[242:]
    data = np.asarray(data["Close"], dtype=float)
    return torch.from_numpy(data)
 


#Time series of our data as a pytorch tensor. 
ada_data = make_tensor("ADA")
btc_data = make_tensor("BTC")
doge_data = make_tensor("DOGE")
eth_data = make_tensor("ETH")

print(btc_data.mean())
print(btc_data.std())
btc_data = btc_data - btc_data.mean()
btc_data = btc_data/btc_data.std()
#plt.plot(ada_data)
#plt.plot(btc_data)
plt.plot(btc_data)
```
